```python
country_codes = []
Countries = []
country_mapping = {}
for item in country_data:
    country_codes.append(item['code'].lower())
    Countries.append(item['name'])
    country_mapping[item['code'].lower()] = item['name']
```

```python
def get_price(url, lst, lst1, price_hist, key,  countries_df, purity_df):
    '''
    url : gold price link to scrape
    lst : prices list to save the prices harvested from the link
    lst1 : Dates list to save the Dates of the price
    price_hist : A dictionary to save the price history photo link
    key : A key for the dictionary to save the photo link
    countries_df : A list to save the country of the iteration for the dataframe
    purity_df : A list to save the purity of the iteration for the dataframe
    '''
    resp = requests.get(url)
    if resp.status_code != 200:
        return
    countries_df.append(key[:2])
    purity_df.append(key[2:])

    soup = BeautifulSoup(resp.text)

    # Loops to save the price and date of the link
    for line in soup.find_all('div', attrs = {'class':'col c9 ccenter'}):
        for txt in range(len(line.text.split())):
            if 'Price:' == line.text.split()[txt]:
                lst.append(line.text.split()[txt+1])
            if 'Updated:' == line.text.split()[txt]:
                lst1.append(line.text.split()[txt+1])

    # Loop to save the price history photo link in th edictionary
    price_hist[key] = None
    for src in soup.findAll('img'):
        if 'gold-price' in src.get('src'):
            price_hist[key] = src.get('src')
```

```python
prices = []
dates = []
price_hist = {}
purity = [24, 23, 22, 21, 18, 14, 12, 10]
purity_df = []
countries_df = []
for i in country_codes:
    for j in purity:
        link = f"https://goldpricez.com/{i}/{j}k/gram"
        get_price(link, prices, dates, price_hist, f'{i+str(j)}', countries_df, purity_df)
```

```python
hist_list = []
for item in price_hist:
    hist_list.append(price_hist[item])
```

```python
df = pd.DataFrame({'Prices in Local Currency' : prices, 'Price History': hist_list, 'Price Date': dates, 'Purity': purity_df,  'Country': countries_df})
df.sample(5)
```

```python
df['Country'] = df['Country'].replace(country_mapping)
df
```

```python
def convert_to_usd(row):
    price_str = row['Prices in Local Currency'].replace('$', '').replace('£', '').replace('৳', '').replace('¥','').replace('฿','').replace('€', '').replace('₩','').replace('₹','0').replace('₦','').replace('₱','').replace(',','')
    price_numeric = float(price_str)
    return price_numeric * exchange_rates.get(row['Country'], 1.0)
```

```python
def convert_to_EGP(row):
    return 30.92*row['PriceUSD']  # From the same source 1 USD = 30.92 EGP
```

```python
df['PriceUSD'] = df.apply(convert_to_usd, axis=1)
df['Price EGP'] = df.apply(convert_to_EGP, axis=1)
df
```
